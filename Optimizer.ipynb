```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#gradint descent
#function : x**4 - 40*x**3 + 2
gradf = lambda x:4*x**3 -120*x**2
lr = 0.00005
cur_x = -20.
iter_cnt = 0
precise = 0.01
max_iter = 100
cur_precise = 1.
ans_x = 30.
x_history = [cur_x]

while cur_precise > precise and iter_cnt < max_iter:
    iter_cnt=iter_cnt +1
    cur_x = cur_x - lr * gradf(cur_x)
    x_history.append(cur_x)
    cur_precise = np.abs(cur_x - ans_x)

x_history = np.array(x_history)
y_history = x_history**4 - 40*x_history**3 + 2

x = np.arange(-20, 40, 0.1)
y = x**4 - 40*x**3 + 2
plt.plot(x, y)
plt.plot(x_history,y_history, 'ro')
print('iter_cnt = {}'.format(iter_cnt))
print('cur_x = {}'.format(cur_x))
```

```python
#momentom
#function : x**4 - 40*x**3 + 2
gradf = lambda x:4*x**3 -120*x**2
lr = 0.00005
cur_x = -20.
iter_cnt = 0
precise = 0.01
max_iter = 100
cur_precise = 1.
ans_x = 30.
x_history = [cur_x]
m = 0.9
v_grad = 0.
while cur_precise > 0.001 and iter_cnt < max_iter:
    iter_cnt=iter_cnt +1
    v_grad = m*v_grad + lr * gradf(cur_x)
    
    cur_x = cur_x - v_grad
    x_history.append(cur_x)
    cur_precise = np.abs(cur_x - ans_x)

x_history = np.array(x_history)
y_history = x_history**4 - 40*x_history**3 + 2

x = np.arange(-20, 40, 0.1)
y = x**4 - 40*x**3 + 2
plt.plot(x, y)
plt.plot(x_history,y_history, 'ro')
print('iter_cnt = {}'.format(iter_cnt))
print('cur_x = {}'.format(cur_x))
```

```python
#Adagrad
#function : x**4 - 40*x**3 + 2
gradf = lambda x:4*x**3 -120*x**2
lr = 1
cur_x = -20.
iter_cnt = 0
precise = 0.01
max_iter = 10000
cur_precise = 1.
ans_x = 30.
x_history = [cur_x]
accu_grad = 0.

while cur_precise > precise and iter_cnt < max_iter:
    iter_cnt=iter_cnt +1
    accu_grad = accu_grad + gradf(cur_x) ** 2
    cur_x = cur_x - (lr / np.sqrt(accu_grad))* gradf(cur_x)
    x_history.append(cur_x)
    cur_precise = np.abs(cur_x - ans_x)

x_history = np.array(x_history)
y_history = x_history**4 - 40*x_history**3 + 2

x = np.arange(-20, 40, 0.1)
y = x**4 - 40*x**3 + 2
plt.plot(x, y)
plt.plot(x_history,y_history, 'ro')
print('iter_cnt = {}'.format(iter_cnt))
print('cur_x = {}'.format(cur_x))
```

```python
#RMSProp
#function : x**4 - 40*x**3 + 2
gradf = lambda x:4*x**3 -120*x**2
lr = 1
cur_x = -20.
iter_cnt = 0
precise = 0.01
max_iter = 10000
cur_precise = 1.
ans_x = 0.
x_history = [cur_x]
accu_grad = 0.
epsilon = 0.9

while cur_precise > precise  and iter_cnt < max_iter:
    iter_cnt=iter_cnt +1
    accu_grad = epsilon * accu_grad + (1-epsilon)*gradf(cur_x)**2
    cur_x = cur_x - (lr / np.sqrt(accu_grad))* gradf(cur_x)
    x_history.append(cur_x)
    cur_precise = np.abs(cur_x - ans_x)

x_history = np.array(x_history)
y_history = x_history**4 - 40*x_history**3 + 2

x = np.arange(-20, 40, 0.1)
y = x**4 - 40*x**3 + 2
plt.plot(x, y)
plt.plot(x_history,y_history, 'ro')
print('iter_cnt = {}'.format(iter_cnt))
print('cur_x = {}'.format(cur_x))
```

```python
#Adam
#function : x**4 - 40*x**3 + 2
gradf = lambda x:4*x**3 -120*x**2
lr = 0.01
cur_x = -20.
iter_cnt = 0
precise = 0.01
max_iter = 10000
cur_precise = 1.
ans_x = 30.
x_history = [cur_x]
accu_grad = 0.
v_grad = 0.
m = 0.9
epsilon = 0.9

while cur_precise > precise  and iter_cnt < max_iter:
    iter_cnt=iter_cnt +1
    accu_grad = epsilon * accu_grad + (1-epsilon)*gradf(cur_x)**2
    v_grad = m*v_grad + (1-m)* gradf(cur_x)
    
    
    cur_x = cur_x - (lr / np.sqrt(accu_grad)/(1-epsilon))* v_grad/(1-m)
    x_history.append(cur_x)
    cur_precise = np.abs(cur_x - ans_x)

x_history = np.array(x_history)
y_history = x_history**4 - 40*x_history**3 + 2

x = np.arange(-20, 40, 0.1)
y = x**4 - 40*x**3 + 2
plt.plot(x, y)
plt.plot(x_history,y_history, 'ro')
print('iter_cnt = {}'.format(iter_cnt))
print('cur_x = {}'.format(cur_x))
```
